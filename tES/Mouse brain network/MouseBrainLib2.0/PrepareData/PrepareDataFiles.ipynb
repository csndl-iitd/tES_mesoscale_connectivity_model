**Ipsilateral regions (Ipsilateral-Ipsilateral)**

```python
# Append '_ipsi' to each element in the LABELS array
LABELS_ipsi = [label + '_ipsi' for label in LABELS]

# Convert the result back to a NumPy array if needed
LABELS_ipsi = np.array(LABELS_ipsi)

# Filter the DataFrame based on the custom-defined list
filtered_df = df.loc[LABELS_ipsi, LABELS_ipsi]

# Remove "_ipsi" from headers
filtered_df.columns = filtered_df.columns.str.replace('_ipsi', '')

# Convert the first column values to strings and then remove "_ipsi"
filtered_df.index = filtered_df.index.str.replace('_ipsi', '')

# Print the filtered DataFrame
filtered_df
```
